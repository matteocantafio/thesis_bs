```python
#convertire la data in formato YYYY-MM-DD

def convert_date_format(df, column_name):
    # Converte la colonna di numeri in una colonna di stringhe
    df[column_name] = df[column_name].astype(str)
    
    # Aggiunge il carattere "-" per separare l'anno, il mese e il giorno
    df[column_name] = df[column_name].apply(lambda x: x[:4] + '-' + x[4:6] + '-' + x[6:])
    
    return df

convert_date_format(df, "Date")

#converto la colonna datetime e hour in una colonna datetime più utile per plottare i grafici
```

```python
#creo una copia del dataframe per convertire le date e orari in formato datatime

df1=df.copy()

#funzione che converte in formato datatime

def convert_to_datetime(row):
    #converto la colonna "Hour" in interi
    hour = int(row['Hour'])
    
    #se l'ora è 24, la imposto a 0 e aggiungo 1 al giorno nella colonna Date
    if hour == 24:
        hour = 0
        date = pd.to_datetime(row['Date']) + pd.Timedelta(days=1)
    else:
        date = pd.to_datetime(row['Date'])
    
    #creo colonna in formato datetime unendo la data e l'ora
    dt = pd.to_datetime(str(date.date()) + ' ' + str(hour) + ':00:00')
    
    return dt

df1['datetime'] = df1.apply(convert_to_datetime, axis=1)


```

## Scomposizione serie storica

Scomporre una serie storica nei pattern significa identificare e analizzare le varie componenti della serie stessa. Una serie storica è composta da quattro elementi fondamentali: trend, stagionalità, componente ciclica e variazione irregolare.

La scomposizione della serie permette di separare tali componenti e analizzarle in modo separato. In questo modo, si possono individuare e comprendere i pattern o le tendenze specifiche dei dati, come il modo in cui cambiano nel tempo e come sono influenzati da altri fattori.


- Trend - Il trend mostra una direzione generale dei dati della serie temporale in un lungo periodo di tempo. Un trend può essere in aumento (ascendente), in diminuzione (discendente) o orizzontale (stazionario).

- Stagionalità - La stagionalità mostra un trend che si ripete in termini di timing, direzione e ampiezza. Ovvero una serie storica dimostra stagionalità ogni qualvolta vi è un cambio regolare e periodico della media della serie. 

- Componente ciclica - Questi sono trend senza ripetizione precisa in un determinato periodo di tempo. Un ciclo si riferisce al periodo di alti e bassi, espansioni e recessioni di una serie temporale, soprattutto osservati nei cicli economici.

- Variazione Irregolare - Queste sono le fluttuazioni nei dati della serie temporale che diventano evidenti quando sono rimossi i trend e le variazioni cicliche. Queste variazioni sono imprevedibili, erratiche e possono essere o meno casuali.

In altre parole, trend, stagionalità, componente ciclica e variazione irregolare sono quattro elementi che possono essere individuati in una serie temporale. Il trend riflette la direzione generale dei dati della serie nel lungo termine, la stagionalità si riferisce a pattern che si ripetono con una certa regolarità, la componente ciclica è caratterizzata da trend che non si ripetono precisamente in un determinato periodo di tempo, mentre la variazione irregolare rappresenta le fluttuazioni casuali che rimangono quando vengono rimosse le altre componenti.

```python
#creo una copia del dataframe per settare la colonna datetime come indice delle righe
df2=df1.copy()

# Imposta la colonna 'datetime' come indice del dataframe
df2.set_index('datetime', inplace=True)

# Crea una variabile contenente la serie storica dei prezzi
serie = df2['PUN']

# Scomponimento della serie storica utilizzando il metodo decompose di statsmodels dove 24h*7gg 
# è il periodo stagionale della serie storica

decomposizione = sm.tsa.seasonal_decompose(serie, model='additive', period=24*7)

# Estrazione della tendenza, stagionalità e residui dalla decomposizione
trend = decomposizione.trend
seasonality = decomposizione.seasonal
residual = decomposizione.resid
```

```python
# Visualizzazione del grafico del trend
plt.figure(figsize=(16,6))
plt.plot(trend)
plt.title('Trend')
plt.ylabel('Prezzo')
plt.xlabel('Data')
plt.show()

# Visualizzazione del grafico della stagionalità
plt.figure(figsize=(16,6))
plt.plot(seasonality)
plt.title('Seasonality')
plt.ylabel('Prezzo')
plt.xlabel('Data')
plt.show()

# Visualizzazione del grafico dei residui
plt.figure(figsize=(16,6))
plt.plot(residual)
plt.title('Residual')
plt.ylabel('Prezzo')
plt.xlabel('Data')
plt.show()
```

Il grafico della stagionalità mostra una variazione regolare nel corso del tempo che si manifesta con periodo settimanale. Questo tipo di stagionalità è quella che incide di più all'interno del nostro dataset

## Previsione dei prezzi dell'energia

La previsione viene effettuata utilizzando un modello di regressione lineare semplice che utilizza il metodo dei minimi quadrati per approssimare la retta di regressione all'interno della nuvola di punti vista nel precedente grafico a dispersione. Questo modello di regressione lineare possiamo importarlo dalla libreria opensource sklearn che fornisce strumenti per l'analisi di dati ed forecasting.



```python
df1.dtypes
```

Non posso utilizzare il dataframe con una colonna in formato datetime per addestrare il modello, quindi devo creare una nuova colonna dove converto i valori della colonna 'datetime' in un numero in modo tale che il modello di regressione possa riconoscerlo. Per questo utilizzo il metodo timestamp che converte ogni riga dell'oggetto datetime in un timestamp (numero di secondi passati dal 1 gennaio 1970, 00:00:00 UTC)

```python
#creo colonna con le date in formato timestamp
df1['timestamp'] = df1['datetime'].apply(lambda x: x.timestamp())
```

```python

df1.dtypes

```

```python
from sklearn.linear_model import LinearRegression

# Creazione del dataset di previsione con le ultime 24 ore del dataframe inerenti al 31 marzo 2023
df_pred = pd.DataFrame({'timestamp_pred': df1['timestamp'].tail(24)})


# Creazione del modello di regressione lineare
model = LinearRegression()

# Suddivisione dei dati di input (train) e quelli di output (test)

#creo un dataset di train tramite delle variabili che mi permettono di selezionare quante righe voglio
#usare per l'addestramento in funzione del periodo di tempo che voglio considerare

s_train=datetime(2022,11,1)
e_train=datetime(2023,3,30)

new_df=df1[(df1['datetime']>=s_train) & (df1['datetime']<=e_train)]

train_dataX = new_df['timestamp']
train_datay = new_df['PUN']

test_data = df_pred

# Addestramento del modello
model.fit(train_dataX.values.reshape(-1,1), train_datay)

# Effettuazione delle previsioni sulle date del 1 aprile 2023
predictions = model.predict(test_data.values.reshape(-1, 1))

# Aggiunta della colonna PUN_pred al dataframe di previsione
df_pred['PUN_pred'] = predictions

# Conversione della colonna datetime_pred da timestamp a datetime
df_pred['datetime_pred'] = pd.to_datetime(df_pred['timestamp_pred'], unit='s')

# Visualizzazione del dataframe di previsione
print(df_pred)


```

In questo codice, prima di tutto, viene creato un nuovo dataframe df_pred contenente le date del 1 aprile 2023. Viene quindi aggiunta la colonna timestamp al dataframe df_pred, che contiene i valori di data e ora in formato timestamp. Successivamente, viene creato un oggetto di regressione lineare con la libreria sklearn e si divide il dataframe di input df1 in dati di training (con colonne timestamp e PUN) e dati di test (solo timestamp). Il modello viene quindi addestrato sui dati di training e viene effettuata la previsione dei prezzi sull'intera giornata del 31 marzo 2023 utilizzando il metodo predict del modello. Infine, i risultati vengono aggiunti al dataframe di previsione df_pred, la colonna datetime_pred viene convertita in formato datetime e il dataframe viene visualizzato a schermo.

Una delle metriche più utilizzate è l'errore assoluto medio o Mean Square Error, ed è la media della differenza in valore assoluto tra la previsione del modello ed il valore obiettivo. Su un numero di 24 previsioni un valore di 21.92 è un indice di errore molto elevato

```python
#print del Mean Absolute error
from sklearn.metrics import mean_absolute_error , r2_score

y_true=df1['PUN'].tail(24)
y_pred=df_pred['PUN_pred']

print("L'errore assoluto medio è {}".format(mean_absolute_error(y_true,y_pred)))
print('Il coefficeinte di determinazione è {}'.format(r2_score(y_true, y_pred)))
```

Il modello di regressione lineare semplice assume che via sia una relazione di tipo lineare tra le variabili indipendenti e la variabile dipendente, ovvero che il prezzo dell'energia sia una funzione lineare del tempo. 

Tuttavia, come abbiamo notato, il prezzo dell'energia ha un andamento fortemente stagionale, il che conferma che la relazione tra il tempo e il prezzo dell'energia sia molto più complessa.

Questa relazione viene esplicata tramite numerose variabili indipendenti che vengono implementate nei modelli di regressione multivariata, quali ad esempio la Capacità di generazione, la domanda di energia e le condizioni ambientali. In particolare la domanda e le condizioni ambientali contribuiscono in modo significativo alla dinamica dei prezzi dell'elettricità e sono modellate attraverso il consumo di elettricità (Electricity Consumption) E-C e i gradi-giorni di riscaldamento e raffreddamento a cui si accede tramite coefficienti tecnici basati sulle condizioni meteorologiche (variabili HDD e DD)

Come possiamo vedere dai grafici il modello di regressione lineare non è in grado di catturare le stagionalità intrinseche del Prezzo dell'energia, che si manifestano con frequenze a livello orario, giornaliero, settimanale, mensile e perfino annuale.

1) Orario: A livello orario, la domanda di energia elettrica varia in modo significativo durante le 24 ore del giorno, a seconda del periodo in cui siamo. In genere, si osserva una maggiore domanda di energia durante le ore di punta, cioè quelle in cui la maggior parte delle persone è sveglia e svolge le proprie attività quotidiane (ad esempio, dalle 18 alle 22). Questo porta ad un aumento dei prezzi dell'energia durante queste ore.

2) A livello giornaliero, la stagionalità del prezzo dell'energia può essere influenzata anche da fattori esterni al mercato energetico, come le condizioni meteorologiche. Ad esempio, nei giorni molto freddi o molto caldi si può verificare una maggiore domanda di energia per il riscaldamento o il condizionamento dell'aria, che può portare ad un aumento dei prezzi.

3) Settimanale: A livello settimanale, invece, i prezzi dell'energia possono essere influenzati dalla presenza di giorni festivi o feriali. In genere, durante i giorni feriali si osserva una minore domanda di energia, a causa della riduzione delle attività produttive e commerciali. Ciò può portare ad una diminuzione dei prezzi.

4) Mensile: A livello mensile, la stagionalità del prezzo dell'energia può essere influenzata da fattori stagionali come l'arrivo dell'estate o dell'inverno. In generale, durante l'inverno si osserva una maggiore domanda di energia per il riscaldamento, mentre durante l'estate si registra un aumento del consumo di energia per il condizionamento dell'aria.

5) Infine, a livello annuale, la stagionalità del prezzo dell'energia può essere influenzata da fattori come l'offerta e la domanda complessive di energia, l'andamento dei mercati energetici internazionali, le politiche governative e le condizioni geopolitiche dei paesi dai quali si ha un import o export di fonti energetiche. Ad esempio il conflitto in ucraina ha avuto come conseguenze un aumento dei costi dell'energia ed una scarsa reperibilità delle materie prime per produrla. Oppure un minor importo d'introduzione di incentivi alle fonti rinnovabili può portare ad una diminuzione dei prezzi dell'energia, mentre l'aumento delle tasse sulle emissioni di gas serra può portare ad un aumento dei prezzi.
